```python
import numpy as np
import pandas as pd
```

```python
file_path = 'test.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Parse the text data into a list of records
records = []
record = {}

for i, line in enumerate(lines):
    line = line.strip()
    if i % 6 == 0:
        if record:  # save the previous record if it exists
            records.append(record)
        record = {'ID': line}
    elif i % 6 == 1:
        record['Name1'] = line
    elif i % 6 == 2:
        record['Score1'] = line
    elif i % 6 == 3:
        record['Name2'] = line
    elif i % 6 == 4:
        record['Score2'] = line

# Add the last record if it exists
if record:
    records.append(record)

# Create a DataFrame
df = pd.DataFrame(records)

# Save the DataFrame to a CSV file
df.to_csv('output.csv', index=False)

# Display the DataFrame
df
```
